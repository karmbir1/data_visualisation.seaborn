```python
import seaborn as sns
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
iris = sns.load_dataset("iris")
```

```python
iris.head()
```

```python
iris.info()
```

```python
iris.describe(include="object")
```

```python
sns.histplot(iris["petal_length"], bins=20)
plt.show()
```

```python
# kde-plot - kernel density plot
```

```python
sns.kdeplot(iris["petal_length"])
plt.show()
```

```python
sns.scatterplot(x=iris["petal_length"],y=iris["petal_width"])
plt.show()
```

```python
sns.scatterplot(x="petal_length",y="petal_width", data=iris, hue="species")
plt.show()
```

```python
setosa = iris.loc[iris['species'] == 'setosa']
versicolor = iris.loc[iris['species'] == 'versicolor']
virginica = iris.loc[iris['species'] == 'virginica']

plt.scatter(x=setosa['petal_length'], y=setosa['petal_width'], c = 'blue')
plt.scatter(x=versicolor['petal_length'], y=versicolor['petal_width'], c = 'orange')
plt.scatter(x=virginica['petal_length'], y=virginica['petal_width'], c = 'green')

plt.show()
```

```python
sns.jointplot(x="petal_length", y="petal_width", data=iris, kind="reg")
plt.show()
```

```python
sns.jointplot(x="petal_length", y="petal_width", data=iris, hue="species")
plt.show()
```

```python
sns.pairplot(data=iris)
plt.plot()
```

```python
# choosing the right visualisation/plot
```

```python
# What is the right chart to use to find dependency between
# one Continuous Variable and Categorical variable?
```

```python
# 1D viz (univariate)- one variable -- num  (hist, kde, boxplot), categorical-count (countplot)
# 2D viz (bivariate)- two variables -- num-num (scatterplot, jointplot), cat-cat (??), cat-num (countplot, boxplot, violin)
# 2D viz - multiple variables -- pairplot, heatmap
# Multi-dimensional viz (multivariate) -- scatter plot with hue, num-num-categorical, cat-cat-num (barplot with hue), surface plot
```

```python
sns.histplot(iris["sepal_length"], bins=20)
```

```python
sns.boxplot(y="sepal_length", data=iris)
```

```python
sns.countplot(x="species", data=iris)
```

```python
n_setosa = iris[iris['species'] == 'setosa'].shape[0]
n_versicolor = iris[iris['species'] == 'versicolor'].shape[0]
n_virginica = iris[iris['species'] == 'virginica'].shape[0]

data = [n_setosa, n_versicolor, n_virginica]
labels=['setosa','versicolor','virginica']

plt.pie(data,
       labels=labels,
       autopct='%.0f%%') # To show the portions in %ages

plt.show()
```
